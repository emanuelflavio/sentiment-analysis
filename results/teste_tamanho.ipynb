```python
import pandas as pd
from langdetect import detect, DetectorFactory
from langdetect.lang_detect_exception import LangDetectException

# Garante que as detecções sejam reprodutíveis
DetectorFactory.seed = 0

# Leia o arquivo CSV

# Supondo que a coluna com os comentários se chame 'comentario'
comentarios = df['Comentario']
comentarios = comentarios.fillna('')

# Listas para armazenar os resultados
comentarios_ingles = []
comentarios_outros = []

# Detectar o idioma de cada comentário
for comentario in comentarios:
    try:
        if detect(comentario) == 'en':
            comentarios_ingles.append(comentario)
        else:
            comentarios_outros.append(comentario)
    except LangDetectException:
        # Se o idioma não puder ser detectado, considerar como "outros"
        comentarios_outros.append(comentario)

# Crie DataFrames separados para os comentários
df_ingles = df[df['Comentario'].isin(comentarios_ingles)]
df_outros = df[df['Comentario'].isin(comentarios_outros)]

# Salve os DataFrames em arquivos CSV separados
df_ingles.to_csv('comentarios_ingles.csv', index=False)
df_outros.to_csv('comentarios_outros.csv', index=False)
```
