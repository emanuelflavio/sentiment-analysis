```python
import matplotlib.pyplot as plt

# Listas com os dados
modelos_tecnicas = [
    "NON + NB", "NON + RF", "NON + LR", "NON + SVM", 
    "RCE + NB", "RCE + RF", "RCE + LR", "RCE + SVM",
    "STOP + NB", "STOP + RF", "STOP + LR", "STOP + SVM",
    "LEM + NB", "LEM + RF", "LEM + LR", "LEM + SVM",
    "STEM + NB", "STEM + RF", "STEM + LR", "STEM + SVM"
]

precisao = [0.82, 0.78, 0.80, 0.82, 0.82, 0.78, 0.80, 0.82, 0.81, 0.77, 0.80, 0.82, 0.82, 0.77, 0.81, 0.82, 0.82, 0.78, 0.80, 0.82]
recall = [0.69, 0.85, 0.77, 0.67, 0.69, 0.85, 0.77, 0.68, 0.69, 0.84, 0.76, 0.68, 0.68, 0.85, 0.77, 0.67, 0.67, 0.85, 0.75, 0.65]
f1_score = [0.74, 0.80, 0.79, 0.73, 0.74, 0.80, 0.78, 0.73, 0.74, 0.79, 0.78, 0.73, 0.74, 0.79, 0.78, 0.73, 0.73, 0.80, 0.77, 0.71]

# Função para plotar gráfico
def plot_metric(metric_values, metric_name):
    plt.figure(figsize=(12, 6))
    plt.bar(modelos_tecnicas, metric_values, color='skyblue')
    plt.xlabel('Modelo e Técnica')
    plt.ylabel(metric_name)
    plt.title(f'{metric_name} por Modelo e Técnica')
    plt.xticks(rotation=90)
    plt.ylim(0, 1)
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
    plt.savefig('teste.png', format='png')

# Plotando gráficos para cada métrica
plot_metric(precisao, "Precisão")
plot_metric(recall, "Recall")
plot_metric(f1_score, "F1-score")
```
