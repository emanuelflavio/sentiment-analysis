```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys


servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
navegador.get("https://letterboxd.com/film/remember-the-night/reviews/by/activity/")
```

```python
import time
lista = []
while True:
    proximo = navegador.find_element(By.CLASS_NAME, 'next')
    mores = navegador.find_elements(By.CLASS_NAME, 'reveal')
    for i in mores:
        if len(i.text) != 0:
            i.click()
    comentarios = navegador.find_elements(By.CLASS_NAME, 'collapsible-text')
    estrelas = navegador.find_elements(By.CLASS_NAME, 'rating')
    for i in comentarios:
        lista.append(i.text)
        print(i.text)
        print()
    proximo.click()
    time.sleep(10)
    
   
```

```python
estrelas = navegador.find_elements(By.CLASS_NAME, 'rating')
txt = estrelas[1].text
print(estrelas[1].text)
quantidade_meia = txt.count('½') * 0.5
quantidade_estrelas = txt.count('★')
soma = quantidade_estrelas + quantidade_meia
print(soma)
```

```python
import time
from selenium.common.exceptions import NoSuchElementException
lista = []
while True:
    li_coment = navegador.find_elements(By.CLASS_NAME, 'film-detail')
    try:
        proximo = navegador.find_element(By.CLASS_NAME, 'next')
    except:
        navegador.quit()
    mores = navegador.find_elements(By.CLASS_NAME, 'reveal')
    for i in mores:
        if len(i.text) != 0:
            i.click()
    time.sleep(1)
    for i in li_coment:
        soma = 0.0
        comentario = i.find_element(By.CLASS_NAME, 'collapsible-text').text
        try:
            estrela = i.find_element(By.CLASS_NAME, 'rating').text
            quantidade_meia = estrela.count('½') * 0.5
            quantidade_estrelas = estrela.count('★')
            soma = quantidade_estrelas + quantidade_meia
        except NoSuchElementException:
            soma = 0.0
        print(soma)
        print(comentario)
        print()
        tupla = comentario, soma
        lista.append(tupla)
    proximo.click()
    time.sleep(5)
            
    
```

```python
print(len(lista))
```

```python
import csv

nome_arquivo = 'remember-the-night.csv'

# Abrindo o arquivo em modo de escrita
with open(nome_arquivo, mode='w', newline='', encoding='utf-8') as arquivo_csv:
    escritor_csv = csv.writer(arquivo_csv)
    
    # Escrevendo o cabeçalho
    escritor_csv.writerow(['Comentario', 'Nota'])
    
    # Escrevendo os dados
    escritor_csv.writerows(lista)

print(f"Arquivo {nome_arquivo} criado com sucesso!")
```
